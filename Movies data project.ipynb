```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

# Read the data from movies.cvs
df = pd.read_csv(r'C:\Users\Ibrahim\Downloads\movies.csv')
```

```python
df
```

```python
# Check for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('The column {} - has {}% missing data.'.format(col, round(pct_missing*100)))
```

```python
# Check for Data Types in the columns

print(df.dtypes)
```

```python
df.drop_duplicates()
```

```python
# Check for any Outliners in the gross

df.boxplot(column=['gross'])
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Check the Correlation Matrix between all numeric columns

df.corr(method ='pearson') #pearson, kendall, spearman
```

```python
c_matrix = df.corr(method='pearson')
corr_pairs = c_matrix.unstack()

print(corr_pairs)
```

```python
sns.heatmap(c_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
# scatter plot with budget - gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget - Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot budget - gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"green"})
plt.title('Budget - Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# scatter plot with budget - score

plt.scatter(x=df['score'], y=df['gross'])
plt.title('Score - Gross Earnings')
plt.xlabel('Score')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
sns.regplot(x='score', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.title('Score - Gross Earnings')
plt.xlabel('Score')
plt.ylabel('Gross Earnings')
plt.show()
```

```python

#Assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
c_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(c_matrix, annot = True)

plt.title("Correlation matrix")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
c_matrix_ext = df.apply(lambda x: x.factorize()[0]).corr()

correlation_pairs = c_matrix_ext.unstack()

print(correlation_pairs)
```

```python
correlation_pairs_sorted = correlation_pairs.sort_values(kind="quicksort")

print(correlation_pairs_sorted)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

high_corr = correlation_pairs_sorted[abs(correlation_pairs_sorted) > 0.5]

print(high_corr)
```
